YOLO

```python
import torch
import sys
from ultralytics import YOLO

def test_gpu():
    print(f"Python Version: {sys.version.split()[0]}")
    print("-" * 30)

    # 1. Check PyTorch & CUDA
    print("Checking PyTorch & CUDA...")
    cuda_available = torch.cuda.is_available()
    
    if cuda_available:
        gpu_name = torch.cuda.get_device_name(0)
        print(f"✅ SUCCESS: CUDA is available!")
        print(f"   GPU Detected: {gpu_name}")
        print(f"   PyTorch Version: {torch.__version__}")
        print(f"   CUDA Version: {torch.version.cuda}")
    else:
        print("❌ ERROR: CUDA is NOT available. Training will be slow (CPU only).")
        return # Stop here if no GPU

    print("-" * 30)

    # 2. Check YOLOv8 on GPU
    print("Checking YOLOv8 computation on GPU...")
    try:
        # Load a tiny model (it will download automatically if you don't have it)
        model = YOLO("yolov8n.pt") 
        
        # Create a dummy image (black image, 640x640)
        import numpy as np
        dummy_image = np.zeros((640, 640, 3), dtype=np.uint8)

        # Force prediction on GPU (device=0)
        results = model.predict(dummy_image, device=0, verbose=False)
        
        # Check if the model is actually on the GPU
        model_device = next(model.model.parameters()).device
        
        if str(model_device).startswith("cuda"):
            print(f"✅ SUCCESS: YOLOv8 is running on {model_device}")
            print("   Test complete. Your environment is ready for training.")
        else:
            print(f"⚠️ WARNING: YOLO ran, but on {model_device}, not CUDA.")
            
    except Exception as e:
        print(f"❌ ERROR running YOLO: {e}")

if __name__ == "__main__":
    test_gpu()
```
