```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
yolo = pd.read_csv("output_demo/Demo3/YOLO/yolo_detection_log.csv")
ssd = pd.read_csv("output_demo/Demo3/SSD/ssd_detection_log.csv")

# Clean data (IMPORTANT)
yolo = yolo.dropna(subset=["Detection_Class"])
ssd = ssd.dropna(subset=["Detection_Class"])

# Sort by frame
yolo = yolo.sort_values("Frame_ID")
ssd = ssd.sort_values("Frame_ID")

# Create figure with 2 subplots
plt.figure()

# -------- YOLO --------
plt.subplot(2, 1, 1)
classes_yolo = yolo["Detection_Class"].unique()

for c in classes_yolo:
    subset = yolo[yolo["Detection_Class"] == c]
    plt.scatter(subset["Frame_ID"], subset["Confidence_Score"], label=c)

plt.title("YOLO: Detection Confidence per Class")
plt.xlabel("Frame ID")
plt.ylabel("Confidence Score")
plt.legend()

# -------- SSD --------
plt.subplot(2, 1, 2)
classes_ssd = ssd["Detection_Class"].unique()

for c in classes_ssd:
    subset = ssd[ssd["Detection_Class"] == c]
    plt.scatter(subset["Frame_ID"], subset["Confidence_Score"], label=c)

plt.title("SSD: Detection Confidence per Class")
plt.xlabel("Frame ID")
plt.ylabel("Confidence Score")
plt.legend()

plt.tight_layout()
plt.show()
```
